```python

"""
Created on Thu Aug 23 17:52:46 2018
@author: chase.kusterer

"""

"""    
DocString:
    A) Introduction:    
        This game is based on a character and style of a horror game called CLOCK TOWER
        which was depeloped by Human Entertainment. It consists of three stages,
        and also has defined functions for start, win, and lose.
        This game is an escape game that in order to win, the player needs wit
        and sometimes pure luck.
        
        Stage 1: Decide whether or not to enter the house.
        Stage 2: Encounter with SCISSORMAN.
        Stage 3: Escape from SCISSORMAN and save friend's father.
    B) Known Bugs and/or Errors:    None, however, there could be some unconsidered inputs things in UX/UI.

"""

import time
import random
from sys import exit
def start_intro():
    global name
    global friend
    print("\033[1m ESCAPE SCISSORMAN \033[0m\n")
    print("Hello! What's your name?\n")
    name = input(prompt = "Enter name :   ") # Prompt for player's name
    error = 0
    while name.isspace() == True or name == "" : # No blank names
        name = input(prompt = "Enter name :   ")
        error +=1
        if error == 4:
            print("\nStop trying to trick me!\n")
            fail()
    while len(name) > 50:
        print("That name is impossible to call! Please enter a shorter one.")
        name = input(prompt = "Enter name :   ")
        error +=1
        if error == 4:
            print("\nStop trying to trick me!\n")
            fail()
    name = name.strip()
    space = name.count("  ")
    while space > 0:
        name = name.replace("  "," ")
        space = name.count("  ")
    name = name.upper()
    error = 0
    time.sleep(0.2)
    print(f"\nOh! Hello \033[1m{name}\033[0m. Welcome to the game!\n")
    time.sleep(0.4)
    print("What\'s your best friend\'s name?\n")
    friend = input(prompt = "Enter name :   ") # Prompt for friend's name
    while friend.isspace() == True or friend == "":
        friend = input(prompt = "Enter name :   ")
        error +=1
        if error == 4:
            print("\nStop trying to trick me!\n")
            fail()
    while len(friend) > 50:
        print("That name is impossible to call! Please enter a shorter one.")
        friend = input(prompt = "Enter name :   ")
        error +=1
        if error == 4:
            print("\nStop trying to trick me!\n")
            fail()    
    friend = friend.strip()
    space = friend.count("  ")
    while space > 0:
        friend = friend.replace("  "," ")
        space = friend.count("  ")
    friend = friend.upper()
    time.sleep(0.2)
    print(f"\nOh! \033[1m{friend}\033[0m? What a coincidence!?! I have a letter from \033[1m{friend}\033[0m for you! \n")
    time.sleep(1)
    print(f"Well, I'll give it to you whenever you\'re ready,\033[1m {name}\033[0m.\n")
    input('<Press ENTER to continue>\n')
    print(f"""
        
        Dear \033[1m{name}\033[0m,
        
            How are you? I hope you're doing well. And I hope this letter finds you well.
        I have to send a letter because this matter cannot be discussed on the phone or 
        via text messages. My father has been working secretly for the government and 
        now he is missing. I don't know if he has been abducted by the government or not,
        that is why I need to be careful. They could be monitoring our incoming and outgoing
        conversations via phones and internet.
        
            Anyway, I need you to come over to my house to help me figure out how to find him.
        I have never seen him leave the house even on the day that he went missing. His car is 
        still here so he could be trapped somewhere in the 3 forbidden rooms. I also overheard 
        something about a guy called SCISSORMAN and he seems to kill people with huge scissors!
        
            Please come quickly, \033[1m{name}\033[0m!
            
        Your best friend, 
                                                                        
        \033[1m{friend}\033[0m
        
            """)
    
    input('<Press ENTER to continue>\n')
    print("\nLet\'s go! I\'ll drive you!")
    time.sleep(1)
    start_escape()

def start_escape():
    count = 0
    error = 0
    while count < 5:
        print(f"""
        You are now in front of \033[1m{friend}\033[0m\'s house. What would you like to do?
        1) Enter the house to find \033[1m{friend}\033[0m.
        2) Ring the door bell.
        3) Go home.
        """)    
        answer = input(prompt = "Type a number \'1\' \'2\' or \'3\' ")
        time.sleep(0.5)
    
        if answer == "1":
            print("\nThe door is not locked! You went inside...\n")
            time.sleep(1)
            inside_house()
        elif answer == "2":
            count +=1
            if count == 4:
                print("\nYou rang the bell too many times... \nThe door opened and came SCISSORMAN!!! \nHe used his huge pair of scissors to cut your head off. x_x\n")
                time.sleep(1.5)
                fail()
            else:
                print("\nNo one answered the door... However, you heard a scream from inside...@_@ \n===========================================================================\n")
        elif answer == "3":
            print("\n===========================================================================\n")
            fail()
        elif error == 5:
            print(f"\nYou messed up the decision. SCISSORMAN opened the door and cut through your torso. You are dead!\n")
            time.sleep(1.5)
            fail()
        else:
            print("\nInput error. Please try again. \n===========================================================================\n")
            error +=1

def inside_house():
    print(f"""
        You stood inside the house. You heard a scream. Sounded like your friend, \033[1m{friend}\033[0m! 
        You ran to find your friend running towards the door.
            "OH great! \033[1m{name}\033[0m, you're here! I just saw my brother dead in the hallway and heard a door shut!
             It's one of those three forbidden rooms but I don't know which one! We need to find my father!"
        You chose one of the doors to open together with \033[1m{friend}\033[0m.
            """)    
    room_list = ["1","2","3"]
    
    rooms = 2    
    
    while rooms > 0:        
        room = random.choice(room_list)        
        input("<Press ENTER to continue>\n")
        print("\n===========================================================================\n")   
        if room == "1":
            room_list.remove("1")
            print(f"You opened the next door...\n\nThere\'s a man with a huge pair of scissors inside \nand he is coming towards you and \033[1m{friend}\033[0m!\n")
            time.sleep(1.5)
            print("Oh, no! It\'s SCISSORMAN! Will you run or fight?")    
            answer = input("Type a word \'Run\' or \'Fight\' :")    
            answer = answer.lower()
            if answer == "run":
                print(f"\nYou quickly shut the door at SCISSORMAN\'s face \nand \033[1m{friend}\033[0m quickly locked the door with a key.\nGreat! You both escaped his murderous scissors and survived to the next room!")
                rooms -=1
            elif answer == "fight":
                print(f"\nYou found a gun! You shot and killed SCISSORMAN! \nAnd you found \033[1m{friend}\033[0m\'s father inside the room alive!")
                time.sleep(1.5)
                win()
            else:
                print(f"\nYou messed up the decision and SCISSORMAN got to you. You and \033[1m{friend}\033[0m are dead!")
                time.sleep(1.5)
                fail()
        elif room == "2":
            room_list.remove("2")
            print("You opened the next door... The room is empty... Moving on...\n")
            rooms -=1
        elif room == "3":
            room_list.remove("3")
            print(f"You opened the next door...\n\nThere\'s a man with a huge pair of scissors inside \nand he is coming towards you and \033[1m{friend}\033[0m!\n")
            time.sleep(1.5)
            print("Oh, no! It\'s SCISSORMAN approaching fast! \nThe only thing right next to you is a stack of documents.\n Do you want to use the documents or run?")    
            answer = input("Type a word \'Use\' or \'Run\' :")    
            answer = answer.lower()
            if answer == "run":
                print(f"\nYou suddenly turned to run and twisted your ankle!\nSCISSORMAN got to you. You and \033[1m{friend}\033[0m are dead!\n")
                time.sleep(1.5)
                fail()
            elif answer == "use":
                print(f"""
        You grabbed the stack of documents with the amount that your hands can possibly handle and
        swung it at SCISSORMAN in the head! SCISSORMAN fell down unconscious
        Great! You both escaped his murderous scissors and survived to the next room!
                """)
                rooms -=1
            else:
                print(f"\nERROR! You messed up the decision and SCISSORMAN got to you. You and \033[1m{friend}\033[0m\ are dead!\n")
                time.sleep(1.5)
                fail()
                    
    print(f"You\'re both still alive! Now the both of you just need to find \033[1m{friend}\033[0m\'s father \nand try to stay alive!\n")    
    input("<Press enter to continue>\n")  
    face_off()
    
def face_off():
    error = 0
    while error < 5:
        print(f"""
        You opened the door of the last room.
        You found a hidden stairs to the underground. 
        Both you and \033[1m{friend}\033[0m went down the stairs 
        and found \033[1m{friend}\033[0m\'s father lying on the floor, still breathing.
            
                "DAD!"
                "Ugh..."
                "Let us help you stand."
            
        You and \033[1m{friend}\033[0m picked \033[1m{friend}\033[0m\'s father up 
        and a gun fell on the floor...
                """)    
        print("SCISSORMAN FOUND YOU!!! What would you do?") 
        print("1) Shoot SCISSORMAN with the gun.")  
        print("2) Run away!")
        answer = input(prompt = "Type a number \'1\' or \'2\' ")
        if answer == "1":   
            print(f"""
        You shot SCISSORMAN but you missed...
        Unfortunately, the gun ran out of bullets...
        You tried to run away but it was too late...")
        SCISSORMAN got to you and cut you in half...  x_x 
                    RIP \033[1m{name}\033[0m
                """)
            time.sleep(1.5)
            fail()
        elif answer == "2":    
            print(f"""
        You tried to run but \033[1m{friend}\033[0m\'s father stopped you.
        "Don't run. Use the gun. Let's end this."
        You decided to take the gun and shoot SCISSORMAN... and you hit!")
        SCISSORMAN is dead. You survived and you saved \033[1m{friend}\033[0m\'s father!
                """)
            time.sleep(1.5)
            win()
        else:
            print("\nInput error. Please try again. \n===========================================================================\n")
            error +=1
    print(f"\nYou messed up the decision and SCISSORMAN got to you and cut you all in half x_x x_x x_x\n")
    time.sleep(1.5)
    fail()
        
def win():
    print(f"Congratulations \033[1m{name}\033[0m!!!\n           \033[1m  <<<<<<<<YOU WON!!!>>>>>>>>\033[0m \n")
    error = 0
    while error < 5:
        time.sleep(1)
        print("\n                        <<<<REPLAY?>>>>")
        replay = input(prompt = "<Input 'Yes' to restart the game, or 'No' to leave the game.>")
        replay = replay.lower()
        if replay == "yes" or replay == "y":
            print("-------------------------------------------------------------\n")
            start_intro() # This takes us back to the start of the game
        elif replay == "no" or replay == "n":
            break
        else:
            error +=1
    print("Thanks for playing! Good bye!")
    exit
              
def fail():
    print("        \033[1m<<<<<<<<GAME OVER>>>>>>>>\033[0m \n")
    error = 0
    while error < 5:
        time.sleep(1)
        print("\n                        <<<<REPLAY?>>>>")
        replay = input(prompt = "<Input 'Yes' to restart the game, or 'No' to leave the game.>")
        replay = replay.lower()
        if replay == "yes" or replay == "y":
            print("-------------------------------------------------------------\n")
            start_intro() # This takes us back to the start of the game
        elif replay == "no" or replay == "n":
            break
        else:
            error +=1
    print("Thanks for playing! Good bye!")
    exit

start_intro()
```
